### I.a On crit les oracles pour le problme primal d'optimisation :

* **OraclePG** : calcul de la fonction et du gradient,
* **OraclePH** : calcul de la fonction du gradient et du Hessien.

```python
def OraclePG(qc):
    
    ## ---> A COMPLETER
    
    w = q0+(B @ qc)
    # critere
    critere = (1/3)* ((w)@(r * w * (np.absolute(w)))) + (pr @ (Ar @ w))
    # gradient
    gradient = B.T @ (r * w * np.absolute(w)) + B.T @ Ar.T @ pr
    
    ## ---> A COMPLETER
    return critere, gradient
```

```python
def OraclePH(qc):
    
    ## ---> A COMPLETER
    
    w = q0+(B @ qc)
    #critere
    critere = (1/3) * np.dot(w, r*w*np.absolute(w)) + np.dot(pr, Ar @ w)
    #gradient
    gradient = B.T @ (r * w * np.absolute(w)) + B.T @ Ar.T @ pr
    #hessien
    hessien = B.T@np.diag(2*r*(np.absolute(q0 + (B @ qc))))@B
    
    ## ---> A COMPLETER
    
    return critere, gradient, hessien
```

###  On charge les prototypes d'algorithmes d'optimisation :

* **Gradient_F** : algorithme de gradient  pas fixe,
* **Newton_F** : algorithme de Newton  pas fixe.

```python
def Gradient_F(Oracle, x0, alpha0):
    
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables
    
    iter_max = 10000
    gradient_step = alpha0
    threshold = 0.000001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []

    time_start = process_time()
    
    x = x0

    ##### Boucle sur les iterations

    for k in range(iter_max):
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)

        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break

        # Direction de descente
        direction = -gradient
        
        # Mise a jour des variables
        x = x + (gradient_step*direction)
        
        # Evolution du gradient, du pas, et du critere
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
    
    ##### Resultats de l'optimisation

    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start
    
    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
    
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

```python
def Newton_F(Oracle, x0, alpha0):
    
    from numpy.linalg import norm
    from numpy.linalg import solve
    from time import process_time
    
    ##### Initialisation des variables    
    
    iter_max = 100
    gradient_step = alpha0
    threshold = 0.000001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []

    time_start = process_time()
    
    x = x0

    ##### Boucle sur les iterations
    
    for k in range(iter_max):
        
        # Valeur du critere et du gradient
        critere, gradient, hessien = Oracle(x)

        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
        
        # Direction de descente
        direction = - solve(hessien, gradient)
        
        # Mise a jour des variables
        x = x + (gradient_step*direction)
        
        # Evolution du gradient, du pas, et du critere
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
   
    ##### Resultats de l'optimisation
    
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
    
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

```python
def BFGS_f(Oracle, x0, alpha0):
    
    from numpy import dot
    from numpy import eye
    from numpy import outer
    from numpy import vdot
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables

    iter_max = 50000
    default_gradient_step = alpha0
    threshold = 0.000001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []
    
    time_start = process_time()

    x = x0

    ##### Boucle sur les iterations
    
    for k in range(iter_max):
 
        # ---> A COMPLETER
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)
        
        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
            
        dim = x.shape[0]       
        # Direction de descente
        if k == 0:
            W_m = np.identity(dim)
            direction = -W_m @ gradient
        else:
            d_x = (x - x_pre)
            d_g = (gradient - pre_gradient)
            
            d = 1/ np.dot(d_g, d_x)
            
            
            multi_x = np.identity(dim) - d*outer(d_x,d_g)
            
            multi_g = np.identity(dim) - d*outer(d_g, d_x)
            
            pre_W_m = W_m
            
            W_m = multi_x @ pre_W_m @ multi_g +  d*outer(d_x , d_x)
            
            direction = -W_m @ gradient
            
        
        
        #wolf algorithm 
        
        x_pre = x
        
        x = x + (default_gradient_step * direction)
        
        
        pre_gradient = gradient
        
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(default_gradient_step)
        critere_list.append(critere)
        
        
        
        # ---> A COMPLETER
        #
        # Attention a l'initialisation de l'algorithme !
    
    # Resultat de l'optimisation
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
        
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

# Variable Size (conditions de Wolfe).m

```python
def Wolfe(alpha, x, D, Oracle):
   
    from numpy import dot
    from numpy import inf
    from numpy.linalg import norm

    ##### Coefficients de la recherche lineaire
   
    omega_1 = 0.1
    omega_2 = 0.9
   
    alpha_min = 0
    alpha_max = inf
   
    ok = 0
    dltx = 0.00000001

    ##### Algorithme de Fletcher-Lemarechal
   
    # Appel de l'oracle au point initial
    argout = Oracle(x)
    critere = argout[0]
    gradient = argout[1]
   
    # Initialisation de l'algorithme
    alpha_n = alpha
    xn = x
   
    # Boucle de calcul du pas
    # xn represente le point pour la valeur courante du pas,
    # xp represente le point pour la valeur precedente du pas.
    while ok == 0:
       
        # Test des conditions de Wolfe
        #
        # - Si les deux conditions de Wolfe sont verifiees,
        #   poser ok = 1 : on sort alors de la boucle while
        # - Sinon, modifier la valeur de alphan et reboucler
        #
        # ---> A COMPLETER
        
        # Point precedent pour tester l'indistinguabilite
        xp = xn
        
        # Point actuel
        xn = x + alpha_n*D
        argout = Oracle(xn)
        critere_n = argout[0]
        gradient_n = argout[1]
        
        #critere_n, gradient_n = Oracle(xn)
        
        if (critere_n > critere + omega_1*alpha_n*dot(gradient,D)):
            alpha_max = alpha_n
            alpha_n = 0.5*(alpha_min + alpha_max)
        elif (dot(gradient_n,D) < omega_2*dot(gradient,D)):
            alpha_min = alpha_n
            if alpha_max == inf:
                alpha_n = 2*alpha_min
            else:
                alpha_n = 0.5*(alpha_min + alpha_max)
        else:
            ok = 1
        
        # ---> A COMPLETER
       
        # Test d'indistinguabilite
        if (norm(xn - xp) < dltx) :
            ok = 2

    return alpha_n, ok
```

```python
def Gradient_V(Oracle, x0, alpha0):
    
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables

    iter_max = 10000
    default_gradient_step = alpha0
    threshold = 0.000001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []
    
    time_start = process_time()

    x = x0

    ##### Boucle sur les iterations

    for k in range(iter_max):
        
        # ---> A COMPLETER
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)
        
        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
            
        # Direction de descente
        direction = -gradient
        
        
        
        #wolf algorithm 
        arg_out = Wolfe(default_gradient_step, x, direction, Oracle)
        gradient_step = arg_out[0]
        
        
        x = x + gradient_step * direction
        
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
        
        # ---> A COMPLETER
    
    ##### Resultats de l'optimisation
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
    
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

# Python fmin Function

```python
from scipy import optimize
```

```python
def f(alpha, *args):
    step = alpha
    x, direction = args
    qc = x + step*direction
    w = q0+(B @ qc)
    return ((1/3)* ((w)@(r * w * (np.absolute(w)))) + (pr @ (Ar @ w)))
```

```python
def Gradient_V_fmin(Oracle, x0, alpha0):
    
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables

    iter_max = 10000
    default_gradient_step = alpha0
    threshold = 0.00001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []
    
    time_start = process_time()

    x = x0

    ##### Boucle sur les iterations

    for k in range(iter_max):
        
        # ---> A COMPLETER
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)
        
        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
            
        # Direction de descente
        direction = -gradient
        
        
        
        #wolf algorithm 
        arg_out = optimize.fmin(f, default_gradient_step,  args=(x, direction)) 
        gradient_step = arg_out[0]
        
        
        x = x + gradient_step * direction
        
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
        
        # ---> A COMPLETER
    
    ##### Resultats de l'optimisation
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
    
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

### I.e On crit l'algorithme de quasi-Newton (BFGS) et on le teste


```python
def BFGS_fmin(Oracle, x0, alpha0):
    
    from numpy import dot
    from numpy import eye
    from numpy import outer
    from numpy import vdot
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables

    iter_max = 250
    default_gradient_step = alpha0
    threshold = 0.000001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []
    
    time_start = process_time()

    x = x0

    ##### Boucle sur les iterations
    
    for k in range(iter_max):
 
        # ---> A COMPLETER
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)
        
        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
            
        dim = x.shape[0]       
        # Direction de descente
        if k == 0:
            W_m = np.identity(dim)
            direction = -W_m @ gradient
        else:
            d_x = (x - x_pre)
            d_g = (gradient - pre_gradient)
            
            d = 1/ np.dot(d_g, d_x)
            
            
            multi_x = np.identity(dim) - d*outer(d_x,d_g)
            
            multi_g = np.identity(dim) - d*outer(d_g, d_x)
            
            pre_W_m = W_m
            
            W_m = multi_x @ pre_W_m @ multi_g +  d*outer(d_x , d_x)
            
            direction = -W_m @ gradient
            
        
        
        #wolf algorithm 
        arg_out = optimize.fmin(f, default_gradient_step,  args=(x, direction))
        gradient_step = arg_out[0]
        
        x_pre = x
        
        x = x + (gradient_step * direction)
        
        
        pre_gradient = gradient
        
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
        
        
        
        # ---> A COMPLETER
        #
        # Attention a l'initialisation de l'algorithme !
    
    # Resultat de l'optimisation
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
        
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```

# Python fmin_bfgs Function

```python
def gradf(alpha, *args):
    import numpy as np
    step = step = alpha
    x, direction = args
    qc = x + step*direction
    w = q0+(B @ qc)
    return (np.dot(direction ,(B.T @ (r * w * np.absolute(w)) + B.T @ Ar.T @ pr)))
```

```python
def Gradient_V_fbfgs(Oracle, x0, alpha0):
    
    from numpy.linalg import norm
    from time import process_time
    
    ##### Initialisation des variables

    iter_max = 10000
    default_gradient_step = alpha0
    threshold = 0.0001
    
    gradient_norm_list = []
    gradient_step_list = []
    critere_list = []
    
    time_start = process_time()

    x = x0

    ##### Boucle sur les iterations

    for k in range(iter_max):
        
        # ---> A COMPLETER
        
        # Valeur du critere et du gradient
        critere, gradient = Oracle(x)
        
        # Test de convergence
        gradient_norm = norm(gradient)
        if gradient_norm <= threshold:
            break
            
        # Direction de descente
        direction = -gradient
        
        
        
        #wolf algorithm 
        arg_out = optimize.fmin_bfgs(f, default_gradient_step,fprime = gradf, args=(x, direction)) 
        gradient_step = arg_out[0]
        
        
        x = x + gradient_step * direction
        
        gradient_norm_list.append(gradient_norm)
        gradient_step_list.append(gradient_step)
        critere_list.append(critere)
        
        # ---> A COMPLETER
    
    ##### Resultats de l'optimisation
    critere_opt = critere
    gradient_opt = gradient
    x_opt = x
    time_cpu = process_time() - time_start

    print()
    print('Iteration :', k)
    print('Temps CPU :', time_cpu)
    print('Critere optimal :', critere_opt)
    print('Norme du gradient :', norm(gradient_opt))
    
    # Visualisation de la convergence
    Visualg(gradient_norm_list, gradient_step_list, critere_list)
    
    return critere_opt, gradient_opt, x_opt
```
